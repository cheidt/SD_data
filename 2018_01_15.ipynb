```python
# print('Number of correct responses: {}')

for code in codes:
    print(code)
    base = dfwt['disp_determinant/medical_code']
    print('Code {}'.format(code))
    code_size = len(base[base==code])
    nem_code_size = len(base[(base==code) & (dfwt['transport_code']!=10)])
    em_code_size = len(base[(base==code) & (dfwt['transport_code']==10)])
    if em_code_size > 0:
        print('Percent level 10 responses: {0:.3f}%'.format(em_code_size/code_size*100))
        over  = len(base[(base==code) & (dfwt['trans_score']>65) & (dfwt['transport_code']==10)])
        under = len(base[(base==code) & (dfwt['trans_score']<65) & (dfwt['transport_code']==10)])
        print('Percent correct responses: {0:.3f}%'.format(over/em_code_size*100))
        print('Percent incorrect responses: {0:.3f}%'.format(under/em_code_size*100))
    print('')
```
